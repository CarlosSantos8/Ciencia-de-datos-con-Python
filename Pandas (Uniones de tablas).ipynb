## Concatenación

Supongamos que tenemos dos dataframes:

```python
Info_Personal = pd.DataFrame({
    "Nombre":["Carlos","Adriana","Daniela","Paulina","Diego"],
    "ApellidoPaterno":["Santos","Pérez","Ochoa","Santiago","Reyes"],
    "ApellidoMaterno":["Alonzo","Rodriguez","Santiago","Díaz","Hidalgo"],
    "Edad":["23","19","20","21","25"],
})

Info_Personal
```

```python
Educacion =  pd.DataFrame({
    "Nombre":["Carlos","Adriana","Daniela","Paulina","Diego"],
    "ApellidoPaterno":["Santos","Pérez","Ochoa","Santiago","Reyes"],
    "ApellidoMaterno":["Alonzo","Rodriguez","Santiago","Díaz","Hidalgo"],
    "Escolaridad":["Universidad","Preparatoria","Universidad","Universidad","Universidad"],
})
Educacion
```

```python
Profesion =  pd.DataFrame({
    "Nombre":["Carlos","Adriana","Daniela","Paulina","Diego"],
    "ApellidoPaterno":["Santos","Pérez","Ochoa","Santiago","Reyes"],
    "ApellidoMaterno":["Alonzo","Rodriguez","Santiago","Díaz","Hidalgo"],
    "Trabajo":["Ingeniero","Estudiante","Abogada","Maestra","Arquitecto"],
})
Profesion
```

```python
Concatenado_Vertical = pd.concat([Info_Personal,Educacion,Profesion],axis = 0,keys = ["Info_Personal","Educación","Profesion"])
Concatenado_Vertical
```

```python
Concatenado_Vertical.loc["Educación"] #Con esto, obtengo el índice que tiene 'Educación'
```

```python
Concatenado_Vertical.loc["Educación"].iloc[1] #Obtengo la primer columna del 2 índice (Educación)
```

```python
Concatenado_Horizontal = pd.concat([Info_Personal,Educacion,Profesion],axis = 1,keys = ["Info_Personal","Educación","Profesion"])
Concatenado_Horizontal
```

Ahora, podremos observar, que si renombramos el índice del DataFrame de ´Profesión´, cambiará nuestra tabla de Concatenado Horizontal

```python
Profesion.index = [0,3,4,7,9]
Profesion
```

```python
Concatenado_Horizontal = pd.concat([Info_Personal,Educacion,Profesion],axis = 1,keys = ["Info_Personal","Educación","Profesion"])
Concatenado_Horizontal
```

Lo anterior sucede, ya que al momento de hacer el <a style="color:red;"> <b> concatenado horizontal </b> </a> , lo que hace es:
Comparar lo del índice 1 del Data Frame <a style="color:blue;"> <b> Info_Personal </b> </a> con lo que tiene el índice 1 del Data Frame <a style="color:blue;"> <b> Educación</b> </a> con lo que tiene el índice 1 del Data Frame <a style="color:blue;"> <b> Profesion </b> </a>.

Observemos que en este último Data Frame, NO tenemos índice 1, por ende lo que hace python es 'rellenar' con NaN

# Joins

Veremos ahora una forma muy importante de relacionar tablas en una base de datos.

<img src="joins.png" width="400">

```python
A = pd.DataFrame({
    "Id_Nombre":["Carlos","Roberto","Alonso","Santos","Daniela"],
    "Correo":["817@gmail.com","rober@gmail.com","Alons827@gmail.com","Santos871hs@gmail.com","Dani37@gmail.com"],
    "Escolaridad":["Universidad","Doctorado","Preparatoria","Maestría","Universidad"]
    })
A
```

```python
B = pd.DataFrame({
    "Nombre":["Carlos","Adriana","Roberto","Alonso","Santos","Carlos"],
    "Estado":["EdoMex","Hidalgo","Guerrero","Sonora","Baja California","CDMX"],
    "EstadoCivil":["soltero","casada","soltero","casado","casado","Casado"]
    })
B
```

```python
InnerJoin = A.join(B.set_index(["Nombre"]),on = ["Id_Nombre"],how = "inner")
InnerJoin #En este caso, lo que obtendremos es la intersección de los 2 conjuntos (Daniela pertenece a A pero no pertenece a B)
```

```python
LeftJoin = A.join(B.set_index(["Nombre"]),on = ["Id_Nombre"],how = "left")
LeftJoin #En este caso, lo que se obtiene es A y la intersección A y B, en este caso Daniela pertenece a A
```

```python
RightJoin = A.join(B.set_index(["Nombre"]),on = ["Id_Nombre"],how = "right")
RightJoin #En este caso, lo que se obtiene es B y la intersección A y B, en este caso Adriana pertenece a B 
```

```python
OutherJoin = A.join(B.set_index(["Nombre"]),on = ["Id_Nombre"],how = "outer")
OutherJoin #En este caso, lo que se obtiene es B y la intersección A y B, en este caso Adriana pertenece a B 
```

# Merge 

Los usuarios experimentados de bases de datos relacionales como SQL estarán familiarizados con la terminología utilizada para describir operaciones de unión entre dos estructuras ( DataFrame objetos) similares a tablas SQL. Hay varios casos a considerar que es muy importante entender:

* <a style="color:black;"> <b> Uniones uno a uno:</b> </a> por ejemplo, cuando se unen dos DataFrameobjetos en sus índices (que deben contener valores únicos).

* <a style="color:black;"> <b> Uniones de muchos a uno :</b> </a> por ejemplo, cuando se une un índice (único) a una o más columnas en un archivo DataFrame.

* <a style="color:black;"> <b> Uniones de muchos a muchos :</b> </a> unir columnas sobre columnas.

Vale la pena dedicar algo de tiempo a comprender el resultado del caso de unión de muchos a muchos . En <a style="color:red;"> <b> SQL/álgebra relacional estándar</b> </a>, si una combinación de teclas aparece más de una vez en ambas tablas, la tabla resultante tendrá el <a style="color:blue;"> <b> producto cartesiano </b> </a>de los datos asociados. A continuación se muestra un ejemplo muy básico con una combinación de teclas única:


```python
A = pd.DataFrame({
    "Nombre":["Carlos","Roberto","Alonso","Santos","Daniela"],
    "Correo":["817@gmail.com","rober@gmail.com","Alons827@gmail.com","Santos871hs@gmail.com","Dani37@gmail.com"],
    "Escolaridad":["Universidad","Doctorado","Preparatoria","Maestría","Universidad"]
    })
A
```

```python
B = pd.DataFrame({
    "Nombre":["Carlos","Adriana","Roberto","Alonso","Santos","Carlos"],
    "Estado":["EdoMex","Hidalgo","Guerrero","Sonora","Baja California","CDMX"],
    "EstadoCivil":["soltero","casada","soltero","casado","casado","Casado"]
    })
B
```

```python
Inner = pd.merge(A, B, on="Nombre") #la función merge, tiene el predeterminado de how='inner'
Inner
```

```python
Inner = pd.merge(A, B, on="Nombre", how='inner') 
Inner
```

```python
Left = pd.merge(A, B, on="Nombre", how='left') 
Left
```

```python
Right = pd.merge(A, B, on="Nombre", how='right') 
Right
```

```python
A = pd.DataFrame(
    {
        "Nombre": ["Carlos", "Carlos", "Roberto", "Camila"],
        "Apellido": ["Santos", "Alonzo", "Santos", "Pérez"],
        "Edad": ["24", "20", "28", "18"],
        "Escolaridad": ["Universidad", "Preparatoria", "Maestría", "Preparatoria"],
    }
)
A
```

```python
B = pd.DataFrame(
    {
        "Nombre": ["Camila", "Carlos", "Carlos", "Daniela"],
        "Apellido": ["Pérez", "Alonzo", "Santos", "Sánchez"],
        "Estado": ["EdoMex", "CDMX", "Hidalgo", "Guerrero"],
        "ClubDeportivo": ["Cruz Azul", "Ámerica", "Tigres", "Tijuana"],
    }
)
B
```

```python
Inner_2Columnas = pd.merge(A, B, on=["Nombre", "Apellido"])
Inner_2Columnas #Intersección de A y B
```

Con lo anterior, obtenemos la intersección de las Columnas 'Nombre' y 'Apellido' de A y B, en este caso, hay 3 registros que aparecen tanto en A como en B.

```python
Left_2Columnas = pd.merge(A, B, how="left",on=["Nombre", "Apellido"])
Left_2Columnas #Conjunto A (Incluye la intersección con B)
```

```python
Right_2Columnas = pd.merge(A, B, how="right",on=["Nombre", "Apellido"])
Right_2Columnas #Conjunto B (Incluye la intersección con A)
```

```python
Outer_2Columnas = pd.merge(A, B, how="outer",on=["Nombre", "Apellido"])
Outer_2Columnas #Es la unión de A y B
```
